```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read all relevant datasets
tweets_hate = pd.read_csv('../data/twitter/all_tweets_hate.csv', 
                          dtype={'id': str,
                                 'positive_score': float, 
                                 'negative_score': float, 
                                 'text': str, 
                                 'hate_value': str})
tweets_sentiment = pd.read_csv('../data/twitter/all_tweets_sentiment.csv', 
                               dtype={'id': str,
                                      'positive_score': float, 
                                      'negative_score': float, 
                                      'text': str, 
                                      'sentiment_value': str})
tweets_perspective = pd.read_csv('../data/twitter/all_tweets_perspective.csv', 
                                 dtype={'id': str, 
                                        'INSULT': float, 
                                        'LIKELY_TO_REJECT': float, 
                                        'IDENTITY_ATTACK': float, 
                                        'SEVERE_TOXICITY': float, 
                                        'THREAT': float, 
                                        'FLIRTATION': float, 
                                        'TOXICITY': float, 
                                        'ATTACK_ON_COMMENTER': float, 
                                        'SEXUALLY_EXPLICIT': float, 
                                        'SPAM': float, #
                                        'INCOHERENT': float, 
                                        'UNSUBSTANTIAL': float, 
                                        'ATTACK_ON_AUTHOR': float, 
                                        'PROFANITY': float, 
                                        'INFLAMMATORY': float, 
                                        'OBSCENE': float})
all_tweets = pd.read_csv('../data/twitter/all_tweets_final.csv', 
                         dtype={'id': str, 
                                'text': str, 
                                'tweet_id': str, 
                                'title': str, 
                                'outlet': str, 
                                'twitter_handle': str, 
                                'article_url': str, 
                                'adfontes_url': str, 
                                'bias_score': float, 
                                'reliability_score': float})
outlet_scores = pd.read_csv('.../data/ad_fontes/outlets_classes_scores.csv',  
                            encoding='cp1252',
                            index_col=0, 
                            dtype={'url': str, 
                                   'bias_class': str, 
                                   'reliability_class': str, 
                                   'reliability_score': float, 
                                   'bias_score': float, 
                                   'outlet': str})
```

```python
# add hate values: join on 'id'
results = all_tweets.merge(tweets_hate.drop(columns=['text']).rename(columns={'positive_score': 'pos_score_hate', 'negative_score': 'neg_score_hate'}), on='id')

# add sentiment values: join on 'id'
results = results.merge(tweets_sentiment.drop(columns=['text']).rename(columns={'positive_score': 'pos_score_sentiment', 'negative_score': 'neg_score_sentiment'}), on='id')

# add perspective scores: join on 'id'
results = results.merge(tweets_perspective, on='id')
```

```python
# add outlet scores & classes (but rename first)
outlet_scores = outlet_scores.rename(columns={'reliability_score': 'overall_reliability', 'bias_score': 'overall_bias'})

results = results.merge(outlet_scores.drop(columns=['outlet']), left_on='adfontes_url', right_on='url').drop(columns=['url'])
```

```python
# get positive and negative polarity, scaled to [0,1]:
sentiment_polarity = results['pos_score_sentiment'].tolist()
```

```python
# get strength of polarity (0.5 is neutral)
for i in range(0, len(sentiment_polarity)):
    sentiment_polarity[i]=abs(sentiment_polarity[i]-0.5)
```

```python
min_value = np.min(sentiment_polarity)
max_value = np.max(sentiment_polarity)
```

```python
# normalize polarity strength to [0, 1]
for i in range(0, len(sentiment_polarity)):
    sentiment_polarity[i]=(sentiment_polarity[i]-min_value) / (max_value-min_value)
```

```python
results['polarity_strength'] = sentiment_polarity
```

```python
# make sure no duplicate tweets are stored:
results = results.drop_duplicates()
```

### PerspectiveAPI Distributions

```python
col_headers = all_tweets.columns.values.tolist()
perspective_api = []

for i in range(0,len(col_headers)):
    if i > 15:
        perspective_api.append(col_headers[i])
    if i > 30:
        break

titles = ['Insult', 'Likely to Reject', 'Identity Attack', 'Severe Toxicity', 'Threat', 'Flirtation', 'Toxicity', 
          'Attack on Commenter', 'Sexually Explicit', 'Spam', 'Incoherent', 'Unsubstantial', 'Attack on Author', 'Profanity',
          'Inflammatory', 'Obscene']
```

```python
for attribute, title in zip(perspective_api, titles):
    sns_plot = sns.distplot(all_tweets[f'{attribute}'], hist=True, kde=False,
                        bins=int(180/5), color = 'blue',
                        hist_kws={'edgecolor':'black'})
    plt.ylabel('Nr. of Tweets')
    plt.xlabel('Score')
    plt.title(f'{title}')
    plt.figure()
    fig = sns_plot.get_figure()
    plt.rcParams.update({'font.size': 13})
    #fig.savefig(f'results/dataset/Perspective API/{title}.png')
```
